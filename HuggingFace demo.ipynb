# Tokenization

```python
from transformers import AutoTokenizer,AutoModelForSequenceClassification,DistilBertTokenizer,DistilBertForSequenceClassification
```

```python

model_name1="nlptown/bert-base-multilingual-uncased-sentiment"
mymodel1=AutoModelForSequenceClassification.from_pretrained(model_name1)
mytokenizer1=AutoTokenizer.from_pretrained(model_name1)

classifier=pipeline("sentiment-analysis",model=mymodel1,tokenizer=mytokenizer1)
res=classifier("I was so not happy with the Barbie Movie")
print(res)
```

```python
tokenizer=AutoTokenizer.from_pretrained("bert-base-uncased")
text="I was so not happy with the Barbie Movie."

tokens=tokenizer.tokenize(text)
print("Tokens:",tokens)
```

```python
# Convert tokens to input IDs
input_ids=tokenizer.convert_tokens_to_ids(tokens)
print("Input IDs:",input_ids)
```

```python
#Encode the text (tokenization + converting to input IDs)
encoded_input=tokenizer(text)
print("Encoded Input:",encoded_input)
```

```python
# Decode the text
decode_output=tokenizer.decode(input_ids)
print("Decode Input :",decode_output)
```
